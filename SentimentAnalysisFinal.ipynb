### Text Data Preprocessing for Tweets and getting sentiment score per tweet

```python
def preprocess(text):
#     if 'via' in text:
#         return re.match('(.*?)via.*?\s?(.*?)', re.sub(r"http\S+", "", re.sub(r"[^a-zA-Z0-9]+", ' ', re.sub(r'\d+', '', text.lower()).translate(str.maketrans('', '', string.punctuation)).strip()))).group(1)
#     else:
        return re.sub(r"http\S+", "", re.sub(r"[^a-zA-Z0-9]+", ' ', re.sub(r'\d+', '', text.lower()).translate(str.maketrans('', '', string.punctuation)).strip()))
    
def get_sentiment(text):
    try:
        res = nlp.annotate(text,
                       properties={'annotators':'sentiment',
            'outputFormat': 'json',
            'timeout': 1000000,
})
        return(res['sentences'][0]['sentiment'],res['sentences'][0]['sentimentValue'])
    except IndexError:
        return None
```

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

### Study the spatial autocorrelation of sentiment score with p value of Moran's I model

```python
pvalue_moran = []
intervals = [
    ['2012-10-22 04:00:00', '2012-10-23 04:00:00'],
    ['2012-10-23 04:00:00', '2012-10-24 04:00:00'],
    ['2012-10-24 04:00:00', '2012-10-25 04:00:00'],
    ['2012-10-25 04:00:00', '2012-10-26 04:00:00'],
    ['2012-10-26 04:00:00', '2012-10-27 04:00:00'],
    ['2012-10-27 04:00:00', '2012-10-28 04:00:00'],
    ['2012-10-28 04:00:00', '2012-10-29 04:00:00'],
    ['2012-10-29 04:00:00', '2012-10-30 04:00:00'],
    ['2012-10-30 04:00:00', '2012-10-31 04:00:00']
]
for x,y in intervals:
    tmp = sandy_geo[(sandy_geo.created_at >= x) & (sandy_geo.created_at <= y)]
    pointInPoly = gpd.sjoin(tmp, puma, how="inner", op='within') 
    pointInPoly = pointInPoly.dissolve(by='GEOID10', aggfunc='mean')
    polygonpuma = pointInPoly.merge(puma, on = "GEOID10")
    polygonpuma.rename(columns = {'geometry_y':'geometry'},inplace = True)
    weights = libpysal.weights.Queen.from_dataframe(polygonpuma)  # generate spatial weights (Queen in this case)
    spatial_auto = esda.Moran(polygonpuma[['score']], weights) 
    mi = spatial_auto
    print(x,y)
    print('Observed I:', mi.I, '\nExpected I:', mi.EI, '\n   p-value:', mi.p_norm)
    pvalue_moran.append(mi.p_norm)
    import geoplot as gplt
    import geoplot.crs as gcrs
```

```python
plt.plot(pvalue_moran,'r--')
plt.ylabel('P value', fontsize = 15)
plt.xlabel('Days', fontsize = 15)
```

### LDA Topics Model

```python
import re
import string
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
import gensim
stop_words = set(stopwords.words('english'))
from nltk.tokenize import word_tokenize
sandy_geo['preprocess'] =sandy_geo['text'].apply(deEmojify).apply(preprocess)
tokens = word_tokenize(str(sandy_geo['preprocess']))
result = [i for i in tokens if not i in stop_words]
lemmatizer=WordNetLemmatizer()
preprocess_text = []
for word in result:
    preprocess_text.append(lemmatizer.lemmatize(word))
pre_token = word_tokenize(str(preprocess_text))
dictionary = gensim.corpora.Dictionary([d.split() for d in preprocess_text])
bow_corpus = [dictionary.doc2bow(doc) for doc in[d.split() for d in preprocess_text]]
lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 10, 
                                   id2word = dictionary,                                    
                                   passes = 10,
                                   workers = 2)
topics = lda_model.print_topics(num_words=6)
for topic in topics:
    print(topic)
```

### Time Series Analysis

```python
import itertools
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(time_sandy,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
mod = sm.tsa.statespace.SARIMAX(time_sandy,
                                order=(1, 0, 0),
                                seasonal_order=(0, 0, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2012-10-31'),end =pd.to_datetime('2012-11-04'),dynamic=False)
pred_ci = pred.conf_int()
ax = time_sandy.plot(label='observed')
pred.predicted_mean.plot(label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Sentiment score')
plt.legend()
plt.show()
```
